Exercice 2 : Analyse graphique basique

```python
dataset_new = dataset.drop(columns = ['Name',  'Ticket', 'PassengerId'])
print('DROP :')
print(dataset_new.head())
```

```python
male = dataset_new[dataset_new['Sex'] == 'male']
female = dataset_new[dataset_new['Sex'] == 'female']
print('Male', male.head(), sep = '\n')
print('Female', female.head(), sep = '\n')
```

```python
# Créer une figure
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Distribution de l'âge pour les hommes survivants et non survivants
sns.histplot(data=male, x='Age', hue='Survived', kde=True, ax=axes[0])
axes[0].set_title('Distribution de l\'âge des hommes')

# Distribution de l'âge pour les femmes survivantes et non survivantes
sns.histplot(data=female, x='Age', hue='Survived', kde=True, ax=axes[1])
axes[1].set_title('Distribution de l\'âge des femmes')

# Afficher les graphiques
plt.tight_layout()
plt.show()
```

existe-il un lien entre les variables ’Embarked’ et ’Survival’

```python
# Tableau croisé entre Embarked et Survived
embarked_survival = pd.crosstab(dataset_new['Embarked'], dataset_new['Survived'])

# Afficher le tableau croisé
print(embarked_survival)

# Représentation graphique
sns.countplot(data=dataset_new, x='Embarked', hue='Survived')
plt.title('Survie en fonction du port d\'embarquement')
plt.show()
```

existe-il un lien entre les variables ’Pclass’ et ’Survival’

```python
# Calculer le tableau croisé avec les proportions par classe
pclass_survival = pd.crosstab(dataset_new['Pclass'], dataset_new['Survived'], normalize='index')
print(pclass_survival.head())
# Renommer les colonnes pour plus de lisibilité
pclass_survival.columns = ['Non Survécu', 'Survécu']

# Afficher le tableau croisé pour visualiser les proportions
print(pclass_survival)

# Représentation graphique à barres empilées des proportions
pclass_survival.plot(kind='bar', stacked=True, figsize=(8, 6), color=['red', 'green'])

# Ajouter un titre et des étiquettes
plt.title('Proportion de survie par classe')
plt.xlabel('Classe')
plt.ylabel('Proportion')
plt.legend(title='Statut de survie')
plt.show()
```
